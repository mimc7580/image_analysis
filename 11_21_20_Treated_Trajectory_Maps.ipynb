## Trajectory maps

```python
#path = '~/Desktop/MPT_Data/Trajectory_Maps/'

ChABC_filename = 'msd_NT_brain_1_slice_1_vid_2'


merged = pd.read_csv(ChABC_filename+'.csv')
particles = int(max(merged['Track_ID']))
ires = 512
figsize = (5,5)
cols = 4
rows = 4

fig = plt.figure(figsize=figsize, dpi=300)
for part in range(0, particles):
    x = merged[merged['Track_ID'] == part]['X']*0.07
    y = merged[merged['Track_ID'] == part]['Y']*0.07
    plt.plot(x, y, alpha=1, color = 'royalblue')

plt.xlim(0, ires*4*0.07)
plt.ylim(0, ires*4*0.07)
#plt.tick_params(axis='both', which='major', labelsize=18, length=6, width=2)
#plt.xticks(np.arange(0,151,25))
#plt.yticks(np.arange(0,151,25))
#plt.ylabel('distance ('r'$\mu$m)', fontsize=20, weight='bold')
#plt.xlabel('distance ('r'$\mu$m)', fontsize=20, weight='bold')
plt.axis('off')
plt.show()
```
